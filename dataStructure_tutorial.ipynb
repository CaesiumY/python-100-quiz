# 초간단 파이썬 - 데이터 분석


## 데이터 분석 패키지 예제

```
import numpy as np
import pandas as pd
import matplotlib

data = np.floor(np.random.rand(50) * 100)

seri = pd.Series(data)

seri.plot()

matplotlib.pyplot.show()
```
