# 73번 최단 경로 찾기

image.png

```
graph = {
    'A': set(['B', 'C',]),
    'B': set(['A', 'D', 'E']),
    'C': set(['A', 'F']),
    'D': set(['B']),
    'E': set(['B', 'F']),
    'F': set(['C', 'E']),
}

def sol():
  start, end = [i for i in input().split(' ')]

  queue = [start]
  visited = [start]
  count = -1

  while queue:
    size = len(queue) # 매 깊이 마다 노드의 숫자를 세어줌
    count += 1 # 다음 깊이로 내려가면 올라감
    print(f'size: {size}')
    print(f'count: {count}')

    for i in range(size): # 현재 깊이의 노드 수만큼 반복함
      node = queue.pop(0)
      if node == end:
        return count
      print(f'node: {node}')
      print(f'graph[node]: {graph[node]}')

    
      for next_node in graph[node]: # 연결된 노드들을 검사
        print(f'next_node: {next_node}')
        if next_node not in visited: # 이미 지나온 노드가 아니면,
          queue.append(next_node) # 다음 while때 비교하기 위해 큐에 넣음
          visited.append(next_node) # 이미 지나왔는지 비교하기 위해 넣음
          print(f'queue: {queue}')
          print(f'visited: {visited}')


    print(f'------------------')
  return count

print(sol())
```

## 프린트 없이 순수하게

```
graph = {
    'A': set(['B', 'C',]),
    'B': set(['A', 'D', 'E']),
    'C': set(['A', 'F']),
    'D': set(['B']),
    'E': set(['B', 'F']),
    'F': set(['C', 'E']),
}

def sol():
  start, end = [i for i in input().split(' ')]

  queue = [start]
  visited = [start]
  count = -1

  while queue:
    size = len(queue)
    count += 1 

    for i in range(size):
      node = queue.pop(0)
      if node == end:
        return count

      for next_node in graph[node]:      
        if next_node not in visited:
          queue.append(next_node)
          visited.append(next_node)
        
  return count

print(sol())
```
