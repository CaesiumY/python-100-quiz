```
graph = {
    'A': set(['B', 'C', 'E']),
    'B': set(['A']),
    'C': set(['A']),
    'D': set(['E', 'F']),
    'E': set(['A', 'D']),
    'F': set(['D']),
}

def bfs(graph, start):
  visited = []
  queue = [start]

  while queue:
    current = queue.pop(0)

    print(f'current: {current}')

    if current not in visited:
      visited.append(current)
      queue += set(graph[current]) - set(visited)
      print(f'visited: {visited}')
      print(f'queue: {queue}')
  
    print(f'---------------')

  return visited


print(bfs(graph, 'E'))
```
