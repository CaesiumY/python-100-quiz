# 74번 최장 경로 찾기

image.png

```
graph = {
    1: [2,3,4],
    2: [1,3,4,5,6],
    3: [1,2,7],
    4: [1,2,5,6],
    5: [2,4,6,7],
    6: [2,4,5,7],
    7: [3,5,6]
}

start, end = [int(i) for i in input().split(' ')]

queue = [start]
visited = []

def solution(n, visited):
  if n[-1] == end: # 재귀 함수를 멈추기 위한 장치
    return len(visited)

  if n[-1] in visited: # 재귀 함수를 멈추기 위한 장치
    return len(visited)

  length = 0
  visited.append(n[-1])

  for next_node in graph[n[-1]]:
    n.append(next_node) # 큐에 넣어서 다음 재귀를 다르게 만들어줌
    length = max(length, solution(n, visited))
    n.pop() # 마지막 도착지점은 뺀다

  return length


print(solution(queue, visited))
```
