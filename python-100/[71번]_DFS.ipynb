```
graph = {
    'A': set(['B', 'C', 'E']),
    'B': set(['A']),
    'C': set(['A']),
    'D': set(['E', 'F']),
    'E': set(['A', 'D']),
    'F': set(['D']),
}

def dfs(graph, start):
  visited = []
  stack = [start]

  while stack:
    current = stack.pop()

    print(f'current: {current}')

    if current not in visited:
      visited.append(current)
      stack += set(graph[current]) - set(visited)
      print(f'visited: {visited}')
      print(f'stack: {stack}')
  
    print(f'---------------')

  return visited


print(dfs(graph, 'E'))
```
