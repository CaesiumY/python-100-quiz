```
def parse_to_minutes(time):
  return int(time.split(':')[0]) * 60 + int(time.split(':')[1])

def parse_to_hour(minutes):
  # return f'{minutes // 60:02d}:{minutes % 60:02d}' :뒤는 포맷 형식, 0은 넣을 문자, 2는 총 길이, d는 10진수 정수형
  return f'{minutes // 60}시간 {minutes % 60}분'

def sol(time_table, current_time):
  result = []
  
  for time in time_table:
    time_diff = parse_to_minutes(time) - parse_to_minutes(current_time)
    if time_diff < 0:
      result.append('지나갔습니다.')
    else:
      result.append(parse_to_hour(time_diff))

  return result


print(sol(["12:30", "13:20", "14:13"], "12:40"))
```
