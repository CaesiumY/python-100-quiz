```
def solution(s):
  length = len(s)
  result = []

  for i in range(1, length + 1):
    for j in range(i):
      print(f'i: {i}, j:{j} - {j}부터 {j+length-i+1}까지는 = {s[j:j+length-i+1]}')
      result.append(s[j:j+length-i+1])

  return result

user_input = input().split(' ')

input_one = set(solution(user_input[0]))
input_two = set(solution(user_input[1]))

answer = input_one.intersection(input_two)
answer = max(answer, key=len)

print(answer)
```
