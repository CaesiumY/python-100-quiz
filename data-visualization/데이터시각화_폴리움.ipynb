```python
import pandas as pd

오름 = pd.read_csv('5.csv', encoding='cp949')

오름.head()
```

```python
for i in range(len(오름)):
  print(오름.iloc[i]['오름명'],오름.iloc[i]['위도'],오름.iloc[i]['경도'])
```

```python
m = folium.Map(location=[33.3729056367668, 126.5319545103804])

tooltip = '클릭해보세요'

for i in range(len(오름)):
  folium.Marker(
    [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
    popup= folium.Popup(f'<strong>{오름.iloc[i]["오름명"]}</strong>', max_width=500),
    tooltip=tooltip,    
  ).add_to(m)

m
```

```python
folium.Circle([33.3729056367668, 126.5319545103804], radius=10000, color='red', fill=True).add_to(m)

m
```

```python
folium.PolyLine(
    locations = [[33.3729056367668, 126.5319545103804],[33.3860915,126.8081632]],
    tooltip='한라산부터 일직선',
).add_to(m)

m
```
